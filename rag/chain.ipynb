```python
#RETRIVER AND cHAIN WITH LANGCHAIN
from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader("paper.pdf")
pdf_docs = loader.load()
pdf_docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=20)
chunk_docs = text_splitter.split_documents(pdf_docs)
chunk_docs
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
db = FAISS.from_documents(chunk_docs, HuggingFaceEmbeddings())
db
```

```python
query = "MRI is a non-invasive method of mapping the internal structure and certain aspects of function within the body."
retrieved_docs = db.similarity_search(query)
retrieved_docs
```
